```python
paragraph = """Dreams are not those which we see while sleeping, but dreams are those when we do not sleep 
before fulfilling them. Dr. Kalam believed that dreams fuel innovation, 
determination, and growth. He often urged the youth to envision a goal so 
compelling that it ignites their passion and drives them toward excellence. 
A true dream is not a fleeting wish but a persistent vision that demands hard work, 
resilience, and sacrifice. He encouraged individuals to break free from the shackles 
of doubt and fear, reminding them that failures are stepping stones to success. 
‘Don’t fear failure,’ he said, ‘as it is part of the journey to greatness.’ 
Kalam’s vision extended beyond personal ambitions to national development, 
emphasizing the collective responsibility of building a self-reliant and 
technologically advanced nation. ‘If you want to shine like a sun, first burn 
like a sun,’ he proclaimed, illustrating the intensity of effort required to 
achieve greatness. His words inspire us to think beyond limitations and act with 
courage, creativity, and conviction, for true dreams lead not only to personal 
fulfillment but also to transformative contributions to society and humanity at 
large."""
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
#stopwords.words('english')
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
sentences = nltk.sent_tokenize(paragraph) # type(sentences) ==> list
```

```python
### Apply Stopwords and Filter and then.... Stemming
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words) # converting all the list of words into sentences
```

```python
# Doing that again with Snowball Stemmer
from nltk.stem import SnowballStemmer
```

```python
snowballstemmer = SnowballStemmer('english')
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i] = ' '.join(words) # converting all the list of words into sentences
```
