```python
# Tokenization Example--------------------------------------------------------------

corpus = """Hello welcome to Rifat's notebook. 
Please do upvote my notebook! Here NLP tokenization is explained in this notebook.
"""
```

```python
print(corpus)
```

```python
# Tokenization:   Paragraph -> Sentence

from nltk.tokenize import sent_tokenize
```

```python
documents = sent_tokenize(corpus)
```

```python
documents # type(documents) => list
```

```python
for sentence in documents:
    print(sentence)
```

```python
# paragraph -> words
# sentences -> words

from nltk.tokenize import word_tokenize
```

```python
word_tokenize(corpus)
```

```python
for sentence in documents:
    print(word_tokenize(sentence))
```

```python
from nltk.tokenize import wordpunct_tokenize
```

```python
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
```

```python
tokenizer = TreebankWordTokenizer()
```

```python
tokenizer.tokenize(corpus)
```

```python
# Stemming--------------------------------------------------------------------------
# eating, eats, eat, eaten ==> (eat) | going, gone, goes ==> (go)
```

```python
words = ["eating", "eats", "eaten", "writing", "programming", "programs", "history", "finally", "finalize"]
```

```python
### PorterStemmer

from nltk.stem import PorterStemmer
```

```python
stemming = PorterStemmer()
```

```python
for word in words:
    print(word + "==>" + stemming.stem(word))
```

```python
### RegxpStemmer class

from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min = 4)
```

```python
reg_stemmer.stem('eating')
```

```python
reg_stemmer.stem('ingeating') # $ing / $ing$
```

```python
### Snowball Stemmer (better than before.......)

from nltk.stem import SnowballStemmer
```

```python
snowballstemmer = SnowballStemmer('english')
```

```python
for word in words:
    print(word + "==>" + snowballstemmer.stem(word))
```

```python
stemming.stem("fairly"), stemming.stem("sportingly")
```

```python
snowballstemmer.stem("fairly"), snowballstemmer.stem("sportingly")
```

```python
# Stemming gives some errors......


### TO SOLVE THAT ==> **LEMMATIZATION** (a dictionary of all words) performes better
#-----------------------------------------------------------------------------------
```

```python
### Lemmatization provides the root word (rather than root stem)
```

```python
### we will be using Wordnet Lemmatizer

from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
lemmatizer.lemmatize("going")
```

```python
words = ["eating", "eats", "eaten", "writing", "programming", "programs", "history", "finally", "finalize"]
```

```python
for word in words:
    print(word + "==>" + lemmatizer.lemmatize(word, pos = 'v'))
```

```python
lemmatizer.lemmatize("goes", pos = 'v')
```

```python
lemmatizer.lemmatize("fairly", pos = 'v'), lemmatizer.lemmatize("sportingly", pos = 'v')
```
