```python
sentence = """Dr. Smith visited Paris last summer and explored landmarks like 
the Eiffel Tower. During her trip, she attended a conference organized by UNESCO 
on June 10, 2024. The event started at 9:00 AM and discussed global education 
policies. Afterward, she purchased a souvenir for $50 and noticed that the local 
tourism industry had seen a 15% growth over the past year."""
```

```python
import nltk
words = nltk.word_tokenize(sentence)
```

```python
tag_elements = nltk.pos_tag(words)
tag_elements
```

```python
### NER ----------------------------------------------------------------------------

nltk.download('maxent_ne_chunker')
nltk.download('words')
```

```python
tree = nltk.ne_chunk(tag_elements)
print(tree.pformat())


# for GUI 
# nltk.ne_chunk(tag_elements).draw()
```
