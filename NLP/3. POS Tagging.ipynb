```python
paragraph = """Good evening, everyone.
We’re at an incredible point in human history. Technology is advancing faster than ever, and the choices we make today will shape the future of humanity for generations.
When I think about the future, I ask myself one fundamental question: How can we accelerate progress toward a better future? To me, that means aiming high and taking risks—whether it’s making life multi-planetary, revolutionizing transportation, or developing sustainable energy solutions.
The truth is, failure is inevitable on the path to success. At SpaceX, we’ve seen rockets explode, and at Tesla, we’ve faced production nightmares. But every failure has taught us something invaluable. Failure is not a step back—it’s a stepping stone.
I urge you to dream big but stay grounded in action. Too often, we underestimate the power of relentless focus. Find something you’re deeply passionate about, something that matters not just to you but to the world, and work harder on it than anyone else.
Remember, progress doesn’t happen by waiting. It happens when people take risks, challenge the status quo, and refuse to accept limits. Each one of you has the power to shape the future.
Let’s build a future we can all be proud of. Let’s aim for the stars while solving the problems here on Earth. Together, there’s nothing we can’t achieve.
Thank you."""
```

```python
import nltk 
sentences = nltk.sent_tokenize(paragraph)
```

```python
sentences
```

```python
### finding the POS Tag-------------------------------------------------------------

from nltk.corpus import stopwords

# Download the stopwords dataset
nltk.download('stopwords')
```

```python
for i in range(len(sentences)):
    words = nltk.word_tokenize(sentences[i])
    words = [word for word in words if word not in set(stopwords.words('english'))]
    # sentences[i] = ' '.join(words)
    pos_tag = nltk.pos_tag(words)

    print(pos_tag)
```
